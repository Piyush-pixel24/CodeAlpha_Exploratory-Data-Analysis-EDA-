```python
# 1. Importing the Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Loading the Dataset
matches = pd.read_csv(r"C:\Users\kumar\Downloads\matches.csv")
```

```python
# Display the First 5 rows
print(matches.head())

# Display the Last 5 rows
print(matches.tail())
```

```python
# Check the dataset shape
print("Shape of Dataset:", matches.shape)
```

```python
# Show the column name
print(matches.columns)
```

```python
#Check Data Types and Dataset Info
print(matches.info())
```

```python
# Check missing values 
print(matches.isnull().sum())
```

```python
# Check Duplicate Rows
print("Duplicate Rows:", matches.duplicated().sum())
```

```python
# Summary Statistics
print(matches.describe())
```

```python
# Mean,Median and Mode on the column GF(Goals For)
# Mean
print("Mean:", matches['gf'].mean())

# Median
print("Median:", matches['gf'].median())

# Mode
print("Mode:")
print(matches['gf'].mode())
```

```python
# Mean Possession
print("Mean:", matches['poss'].mean())

# Median Possession
print("Median:", matches['poss'].median())

# Mode Possession
print(matches['poss'].mode())
```

```python
#Top Teams by Goals Scored
plt.figure(figsize=(12,6))

matches.groupby('team')['gf'].sum().sort_values(
    ascending=False
).head(10).plot(kind='bar')

plt.title("Top Teams by Goals Scored")
plt.xlabel("Teams")
plt.ylabel("Goals Scored")

plt.xticks(rotation=45)

plt.show()
```

```python
#Teams Conceding Most Goals
plt.figure(figsize=(12,6))

matches.groupby('team')['ga'].sum().sort_values(
    ascending=False
).head(10).plot(kind='bar', color='red')

plt.title("Teams Conceding Most Goals")
plt.xlabel("Teams")
plt.ylabel("Goals Conceded")

plt.xticks(rotation=45)

plt.show()
```

```python
# match result distribution
matches['result'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Match Results Distribution")

plt.ylabel("")

plt.show()
```

```python
# possession distribution histogram
plt.figure(figsize=(10,5))

plt.hist(matches['poss'], bins=20)

plt.title("Possession Distribution")
plt.xlabel("Possession %")
plt.ylabel("Frequency")

plt.show()
```

```python
#correlation heatmap
plt.figure(figsize=(12,8))

sns.heatmap(
    matches.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
#Goals For vs Goals Against
plt.figure(figsize=(8,5))

plt.scatter(
    matches['gf'],
    matches['ga']
)

plt.title("Goals For vs Goals Against")
plt.xlabel("Goals For")
plt.ylabel("Goals Against")

plt.show()
```

```python
#teams with higest average possession
plt.figure(figsize=(12,6))

matches.groupby('team')['poss'].mean().sort_values(
    ascending=False
).head(10).plot(kind='bar', color='green')

plt.title("Teams with Highest Average Possession")
plt.xlabel("Teams")
plt.ylabel("Average Possession")

plt.xticks(rotation=45)

plt.show()
```

```python
#Top Captains by Matches Played
plt.figure(figsize=(12,6))

matches['captain'].value_counts().head(10).plot(
    kind='bar',
    color='orange'
)

plt.title("Top Captains by Matches")
plt.xlabel("Captain")
plt.ylabel("Matches")

plt.xticks(rotation=45)

plt.show()
```

```python
# Average Attendance by Team
plt.figure(figsize=(12,6))

matches.groupby('team')['attendance'].mean().sort_values(
    ascending=False
).head(10).plot(kind='bar')

plt.title("Average Attendance by Team")
plt.xlabel("Teams")
plt.ylabel("Attendance")

plt.xticks(rotation=45)

plt.show()
```

```python
# Top Referees Officiating Matches
plt.figure(figsize=(12,6))

matches['referee'].value_counts().head(10).plot(
    kind='bar',
    color='brown'
)

plt.title("Top Referees")
plt.xlabel("Referee")
plt.ylabel("Matches Officiated")

plt.xticks(rotation=45)

plt.show()
```

```python
#Win/Loss Analysis
plt.figure(figsize=(10,5))

matches['result'].value_counts().plot(
    kind='bar',
    color='skyblue'
)

plt.title("Win/Loss/Draw Analysis")
plt.xlabel("Result")
plt.ylabel("Count")

plt.show()
```

```python
# Top Opponent Faced 
plt.figure(figsize=(12,6))

matches['opponent'].value_counts().head(10).plot(
    kind='bar',
    color='purple'
)

plt.title("Top Opponents Faced")
plt.xlabel("Opponent")
plt.ylabel("Matches")

plt.xticks(rotation=45)

plt.show()
```

```python
# Save Cleaned Datasets
matches.to_csv("cleaned_matches.csv", index=False)

print("Cleaned dataset saved successfully.")
```
